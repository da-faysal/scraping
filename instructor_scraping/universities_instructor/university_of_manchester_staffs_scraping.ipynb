```python
# Import required modules
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
```

```python
# We will use selenium to read the table in pandas
driver = webdriver.Chrome(r"/home/faysal/Documents/utilities/chromedriver")
driver.get("https://www.manchester.edu/faculty-staff-directory?f/")
wait = WebDriverWait(driver, 10)
time.sleep(20)

# Read the html table in pandas. This return a list of dataframes
df = pd.read_html(driver.find_element_by_tag_name("table").get_attribute("outerHTML"))

# Convert it to df from list
df = pd.concat(df) # or, df = df[0]

# Let's preview the data and save as csv
df.head(10)
```
