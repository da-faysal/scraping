```python
# Import required modules
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
```

```python
# We will use selenium to read the table in pandas
driver = webdriver.Chrome(r"/home/faysal/Documents/utilities/chromedriver")
driver.get("https://giesbusiness.illinois.edu/faculty-research/faculty-staff-directory")
wait = WebDriverWait(driver, 10)
time.sleep(7)

# Locate using tag name "table" and scrape the content and read in  as pandas dataframe
df = pd.read_html(driver.find_element_by_tag_name("table").get_attribute("outerHTML"))[0]
```

```python
# Extract name and designation from Name
df.insert(loc=0, value= df.Name.str.split(",").str.get(0), column="name")
df.insert(loc=1, value= df.Name.str.split(",").str.get(1), column="designation")

# Extract email and phone from Contact
df.insert(loc=df.shape[1], value=df.Contact.str.split("edu").str.get(0) + str("edu"), column="email")
df.insert(loc=df.shape[1], value=df.Contact.str.split("edu").str.get(-1), column="phone")

# Drop some columns 
df.drop(["Name", "Contact"], axis=1, inplace=True)

# Let's see what we have
df.head(10)
```
