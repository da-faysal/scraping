```python
# Import required modules
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Create cover pages for live online courses
cover_page = []
for pg in range(1, 6):
    cover_page.append(f"https://www.reed.co.uk/courses/?pageno={pg}&keywords=Online%20with%20live%20classes&sortby=MostPopular&pagesize=100")
```

```python
# This function scrapes study method along with other info
def get_study_method(url):
    # Make request
    r = requests.get(url)
    s = BeautifulSoup(r.text, "lxml")
    
    # List of emptry variables to scrape
    course_title = []
    course_link = []
    offer_price = []
    study_method = []
    
    # Main container for a course
    cont = s.find_all(class_="card card-body course d-sm-flex")
    
    for c in cont:
        # Scrape course title
        try:
            course_title.append(c.find("a").text.strip())
        except:
            course_title.append("na")
        
        # Scrape course link
        try:
            course_link.append("https://www.reed.co.uk" + c.find("a").get("href"))
        except:
            course_link.append("na")
        
        # Scrape offer price
        try:
            offer_price.append(c.find(class_="price").text.strip().replace("£", ""))
        except:
            offer_price.append("na")
        
        # Scrape study method
        try:
            study_method.append(c.find(class_="book-icon").text.strip())
        except:
            study_method.append("na")
            
    # Temporary dataframe
    temp_df = pd.DataFrame({
        "course_title":course_title,
        "course_link":course_link,
        "offer_price":offer_price,
        "study_method":study_method
    })
    return temp_df
```

```python
# Apply the function
df = pd.concat(list(map(get_study_method, cover_page))).reset_index(drop=True)
df.head(10)
```
