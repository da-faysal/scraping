```python
# Import required modules
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from datetime import date
```

```python
# Making request to the site and parsing as beautifulsoup object
r = requests.get("https://www.shawacademy.com/courses")
s = BeautifulSoup(r.text, "lxml")

# This is the main container
main_cont = s.find_all(class_="coursecatalogue-module--card--OJhV7")

# Empty list of variables to scrape
course_title = []
course_link = []
unit_sold = []

# Iterate over the main container
for cont in main_cont:
    
    # Scrape course title
    try:
        course_title.append(cont.find(class_="coursecatalogue-module--courseName--24Gl2").text.strip())
    except:
        course_title.append("na")
    
    # Scrape course links
    try:
        course_link.append("https://www.shawacademy.com" + cont.find(class_="coursecatalogue-module--cardList--cNDGb").get("href"))
    except:
        course_link.append("na")
    
    # Scrape unit sold
    try:
        unit_sold.append(cont.find(class_="coursecatalogue-module--enrollCount--1ad8n").text.replace("Graduates", "").replace(",", "").strip())
    except:
        unit_sold.append(0)

        
# Create a df off scraped dataframe
df = pd.DataFrame({
    "course_title":course_title,
    "course_link":course_link,
    "unit_sold":unit_sold
})
```
